```python
from sklearn.datasets import make_circles,make_blobs
from matplotlib import pyplot
import numpy as np
from pandas import DataFrame

X, y = make_circles(n_samples=(40,40), noise=0.07)

t,p = make_blobs(n_samples=45,centers=([(0,0)]),cluster_std=0.25, n_features=2)
p=[2 for i in p]
X=np.concatenate((X, t))
y=np.concatenate((y, p))
X=np.array([[i[0]+3,i[1]*2+4] for i in X])

t,p = make_blobs(n_samples=50,centers=([(4.5,6)]),cluster_std=0.3, n_features=2)
p=[3 for i in p]
X=np.concatenate((X, t))
y=np.concatenate((y, p))

t,p = make_blobs(n_samples=50,centers=([(5,3)]),cluster_std=0.3, n_features=2)
p=[4 for i in p]
X=np.concatenate((X, t))
y=np.concatenate((y, p))

df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
colors = {0:'orange', 1:'blue', 2:'magenta', 3:'green', 4:'yellow'}
fig, ax = pyplot.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y',s=60, label=['Tiger','Lion','Penguin','Dolphin','Python'][key], color=colors[key])
pyplot.xticks(np.arange(2, 6+1, 1.0))
pyplot.yticks(np.arange(2, 7+1, 1.0))

pyplot.legend(loc='upper right')
pyplot.show()
```

```python
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features.
y = iris.target

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5

plt.figure(1, figsize=(8, 6))

# # Plot the training points
a=plt.scatter(X[:, 0][y==0], X[:, 1][y==0], color='Blue', edgecolor='k')
b=plt.scatter(X[:, 0][y==1], X[:, 1][y==1], color='Green', edgecolor='k')
c=plt.scatter(X[:, 0][y==2], X[:, 1][y==2], color='Yellow', edgecolor='k')
plt.legend((a,b,c),('Setosa','Versicolour','Virginica'))
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
iris = datasets.load_iris()
X = iris.data
y = iris.target


# Plot the ground truth
fig = plt.figure(1, figsize=(4, 3))
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

for name, label in [('Setosa', 0),
                    ('Versicolour', 1),
                    ('Virginica', 2)]:
    ax.text3D(X[y == label, 3].mean(),
              X[y == label, 0].mean(),
              X[y == label, 2].mean() + 2, name,
              horizontalalignment='center',
              bbox=dict(alpha=.2, edgecolor='w', facecolor='w'))

y = np.choose(y, [1, 2, 0]).astype(float)
ax.scatter(X[:, 3], X[:, 0], X[:, 2], c=y, edgecolor='k')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_xlabel('Petal width')
ax.set_ylabel('Sepal length')
ax.set_zlabel('Petal length')
ax.set_title('Ground Truth')
ax.dist = 12

fig.show()
pyplot.show()
```
